```python
url="https://api.github.com/repos/pandas-dev/pandas/issues"
pd.read_json(url)
```

```python
#we are requesting for the data and we are getting a response in the exhange of requesting get(basically means ur requesting for the data)
import requests
data=requests.get(url)
```

```python
data
```

```python
#if we want to know what is there in the data
data.json()
```

```python
#the data in the json is stored as a key value pair
```

```python
df=data.json()
len(df)
```

```python
for i in range(len(df)):
  print(i)
```

```python

#for accesing in the dataset
for i in range(len(df)):
  print(df[i])
```

```python
#for accessing users inside of the dataset we use
for i in range(len(df)):
  print(df[i]["user"])
```

```python
#in each user we want something like nodeid present in it
for i in range(len(df)):
  print(df[i]["user"]["node_id"])
```
